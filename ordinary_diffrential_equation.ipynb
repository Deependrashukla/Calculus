```python
import matplotlib.pyplot as px
import plotly.graph_objects as go
import math as m 
```

```python
# for analytical method
t_list = [t/10 for t in range (50)] # contains time
x_list = [m.exp(t) for t in t_list]# analytical value corresponding to t_list
def analytical():
  px.plot(t_list, x_list, label =  'analyticals graph')
  px.legend()
  #px.show()
  
analytical()
```

```python
# for numerical method
y_list = [1] # containing numerical value
def numerical():
  for i in range(1,50):
    x = y_list[i-1]*0.1 + y_list[i-1]# formula of numerical value x(t + dt) = x(t)*dt + x(t)
    y_list.append(x)

  px.plot(t_list, y_list, label = 'graph of numerical method' )
  px.legend()
  
  #px.show()
analytical()
numerical()

```

```python
# for error in numerical corresponding to analytical 

error_lst = []
dt_lst = [i/100 for i in range(1000)]
for dt in dt_lst:
  error = ((m.exp(dt) - (1*dt + 1))) / m.exp(dt)
  error_lst.append(error)
px.plot( dt_lst, error_lst, label = 'numerical error vs delta t')
px.legend()
#px.show()

```

```python
# for midpoint method
def midpoint():
  y_list = [1]
  for i in range(1,50):
    # formula of midpoint value x(t) = x(t) + delta t(x(t) + delta t/2 * x(t))
    x = 0.1*(y_list[i-1] + 0.05*y_list[i-1]) + y_list[i-1]
    y_list.append(x)

  px.plot(t_list, y_list, label = 'midpoints graph')
  px.legend()
  #px.show()
midpoint()
analytical()
```

```python
# for error in midpoint corresponding to analytical 
def midpoint_error():
  global dt_lst
  error_lst = []
  for dt in dt_lst:
    error = ((m.exp(dt) - ( 1*dt + 1*dt/2 + 1 ))) / m.exp(dt)
    error_lst.append(error) 

  px.plot(dt_lst, error_lst, label = 'midpoint error vs delta t')
  px.legend()
  #px.show()
midpoint_error()
```
